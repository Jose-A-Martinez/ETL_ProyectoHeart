```python
import pandas as pd
from scipy.stats import zscore
```

```python
# Crear un DataFrame
df = pd.read_csv(r'C:/Users/marti/OneDrive/Escritorio/Henry/Proyecto_Heart/Heart_Disease_Prediction.csv')
```

```python
# CONTROLAR COLUMNAS SELECCIONADAS 
# Seleccionar solo las columnas numÃ©ricas
num_cols = ['Age', 'BP', 'Cholesterol', 'Max HR', 'ST depression', 'Number of vessels fluro']
```

```python

# Calcular los cuartiles Q1 y Q3 y el rango intercuartil (IQR) solo para las columnas numÃ©ricas
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Identificar los outliers
df_outliers = df[(df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))].dropna(how='all')
df_outliers = df_outliers.align(df, join='right', axis=1)[0]

print(df_outliers)
```

```python
# Filtrar solo los valores outliers
outliers = df_outliers.stack().dropna()

# Filtrar solo los valores outliers de columnas especificas
#outliers = df_outliers['BP'].dropna() 
#outliers = df_outliers.loc[:, ['Cholesterol', 'BP']].stack().dropna() 

# Mostrar solo los valores outliers
print(outliers)
```

```python
# Calcular el puntaje z para cada valor en las columnas numÃ©ricas seleccionadas
z_scores = df[num_cols].apply(zscore)

# Eliminar los valores que tengan un puntaje z mayor a 3 o menor a -3
df_clean = df[(z_scores < 3).all(axis=1) & (z_scores > -3).all(axis=1)]
```
